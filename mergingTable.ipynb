```python
import numpy as np
import pandas as pd
```

```python
#table with EC50-values = pChEMBL-Values (-log10(ChEMBL-Value))
fpr2BindingData = pd.read_csv("fpr2Full.csv",sep = ";", decimal = ",") 

#table with 50% similarity to FUI-ligand
similarStructure = pd.read_csv("40similarStructure.csv", sep = ";", decimal = ",")
```

```python
fpr2BindingData.shape
```

```python
fpr2BindingData.head(50)
```

```python
fpr2BindingData = fpr2BindingData.drop("Standard_Value", axis=1)
```

```python
fpr2BindingData.head(50)
```

```python
fpr2BindingData["pChEMBL_Value"] = pd.to_numeric(fpr2BindingData["pChEMBL_Value"], errors='coerce')
```

```python
fpr2BindingData.head(30)
```

```python
fpr2BindingData = fpr2BindingData.dropna(subset = "pChEMBL_Value")
```

```python
fpr2BindingData.shape
```

```python
fpr2BindingData.head(30)
```

```python
similarStructure = similarStructure.drop("Smiles", axis=1)
```

```python
similarStructure.head()
```

```python
similarStructure = similarStructure.rename(columns={"ChEMBL ID" : "ChEMBLID"})
```

```python
fpr2BindingData.head(20)
```

```python
fpr2BindingData.shape
```

```python
similarStructure.head()
```

```python
similarStructure.shape
```

```python
mergedTable = fpr2BindingData.merge(similarStructure, on="ChEMBLID", how="right") 
```

```python
mergedTable.shape
```

```python
mergedTable.head(10)
```

```python
mergedTable.head()
```

```python
mergedTable = mergedTable.dropna(subset = "pChEMBL_Value")
```

```python
mergedTable.head(10)
```

```python
mergedTable.shape
```

```python
mergedTable = mergedTable.sort_values("pChEMBL_Value", ascending=[False])
```

```python
mergedTable.head(10)
```

```python
mergedTable["activity"] = ["highly_active" if x >= 7.0 else "partial_active" for x in mergedTable["pChEMBL_Value"]]
```

```python
def condition(x):
    if x >= 7.3:
        return "active"
    elif x<5:
        return "inactive"
    else:
        return "partial_active"
    
#Applying the conditions
fpr2BindingData["activity"] = fpr2BindingData["pChEMBL_Value"].apply(condition)
```

```python
fpr2BindingData.shape
```

```python
mergedTable.shape
```

```python
mergedTable.head()
```

```python
mergedTable = mergedTable.drop("pChEMBL_Value", axis=1)
mergedTable = mergedTable.drop("activity", axis=1)
mergedTable = mergedTable.drop("Smiles", axis=1)
```

```python
mergedTable.head()
```

```python
mergedTable = mergedTable.sort_values("Similarity", ascending = False).drop_duplicates("ChEMBLID")
```

```python
mergedTable.shape
```

```python
fpr2bindingSortert = fpr2BindingData.sort_values("pChEMBL_Value", ascending = False).drop_duplicates("ChEMBLID").sort_index()
```

```python
fpr2bindingSortert.head()
```

```python
fpr2bindingSortert.shape
```

```python
mergedTable.head()
```

```python
finalMergedTable = fpr2bindingSortert.merge(mergedTable, on="ChEMBLID", how="left")
```

```python
finalMergedTable = finalMergedTable.sort_values("Similarity", ascending = False)
```

```python
finalMergedTable.shape
```

```python
finalMergedTable = finalMergedTable.drop_duplicates("ChEMBLID")
```
